#### 빅분기 기출문제 풀이 #2 
- Data : https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv 
1. 만족도 컬럼에 결측치가 존재한다. 전체 만족도의 평균값으로 대체하라. 
2. 근속연수의 결측치는 같은 부서내 동일한 성과등급을 가진 직원들의 평균 근속연수로 대체 
3. 각 직원의 연봉 / 근속연수 값을 계산, 그 값이 세번째로 큰 직원의 근속연수를 구하라
4. 각 직원의 연봉 / 만족도 값을 계산, 그 값이 두번째로 큰 직원의 교육 참가횟수를 구하라
5. 3과 4의 값을 더한 결과를 구하라 (정수출력)

```python
df = pd.read_csv('https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv')
print(df.shape)
df.head()
```

```python
# 결측치 확인 
df.isnull().sum()
```

```python
# 만족도 결측치 대체 
df['만족도'] = df['만족도'].fillna(df['만족도'].mean())

# 근속연수 결측치 대체 
res1 = df.groupby(['부서','성과등급'])['근속연수'].transform('mean')
df['근속연수'] = df['근속연수'].fillna(res1)

# 결과확인 
df.isnull().sum()
```

```python
# 연봉/근속연수 계산 
df['ratio1'] = df['연봉']/df['근속연수'] 
df.sort_values(by='ratio1', ascending=False)

# 연봉/만족도 계산 
df['ratio2'] = df['연봉']/df['만족도'] 
df.sort_values(by='ratio2', ascending=False)
```

```python
#Q20. launch time
df = pd.read_csv('data/delivery_time.csv')
#print(df.info())
display(df.head())
print('전체주문건 =', len(df))

df['주문시간'] = pd.to_datetime(df['주문시간'])
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])
df['예상도착시간'] = pd.to_datetime(df['예상도착시간'])
#print(df.info())

df = df[
    (df['주문시간'].dt.hour >= 10) & (df['주문시간'].dt.hour <=13)
]
print('점심주문건 = ', len(df))

df['소요시간'] = (df['실제도착시간'] - df['주문시간']).dt.total_seconds() / 3600
df['배달속도'] = df['거리'] / df['소요시간']
df['과속'] = df['배달속도'] > 40

print('점심시간 과속 건수 = ', df['과속'].sum())
```

```python
#Q21. delivery fee
df = pd.read_csv('data/delivery_time.csv')
#print(df.info())
display(df.head())

df['주문시간'] = pd.to_datetime(df['주문시간'])
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])
df['예상도착시간'] = pd.to_datetime(df['예상도착시간'])
#print(df.info())

def d_fee(x): 
    if x < 5 : 
        return 2000
    elif x < 10: 
        return 4000 
    elif x < 15: 
        return 6000
    else: 
        return 8000

df['배달료'] = df['거리'].apply(d_fee)    

df['yy_mm'] = df['주문시간'].dt.to_period('M')
df.groupby('yy_mm')['배달료'].sum()
```
