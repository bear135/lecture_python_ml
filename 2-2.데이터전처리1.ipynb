## 결측치

```python
# 타이타닉 데이터 불러와 결측치 확인 
df = sns.load_dataset('titanic')
df.isnull()
```

```python
df.isnull().sum()
```

```python
# age의 결측치를 평균나이로 대체 
df['age'] = df['age'].fillna(df['age'].mean())

# deck의 결측치를 최빈값으로 대체 
df['deck'] = df['deck'].fillna(df['deck'].mode()[0])

# embark_town 결측치는 해당 row를 삭제
df = df.dropna(subset=['embark_town'])   

# 결측치 정리 결과 확인 
df.isnull().sum()
```

#### 특정값 변경

```python
df.head()
```

```python
# sex 컬럼의 male = 0, female = 1로 변경 
df2 = df.copy()

df2['sex'] = df2['sex'].replace('male', 0)
df2['sex'] = df2['sex'].replace('female', 1)
```

```python
# map()으로 여러개의 값을 변경 
# class 컬럼의 값 변경 : First -> 1st, Second -> 2nd, Third -> 3rd 
class_map = {
    'First': '1st', 
    'Second': '2nd', 
    'Third': '3rd'
}

df3 = df.copy()
df3['class'] = df3['class'].map(class_map)
```

```python
df4 = df.copy()
df4['class'] = df4['class'].replace(class_map)
df4.head()
```

#### groupby

```python
# flights 데이터 로딩 
flights = sns.load_dataset('flights')
#flights

# 연도x월별 승긱수 합계, 평균, 최소, 최대
flights.groupby(['year', 'month'])['passengers'].sum()
flights.groupby(['year', 'month'])['passengers'].mean()
flights.groupby(['year', 'month'])['passengers'].max()
flights.groupby(['year', 'month'])['passengers'].min()

# agg() 사용 
flights.groupby(['year', 'month'])['passengers'].agg(['sum', 'mean', 'max', 'min'])

# new DataFrame (인덱스 초기화)
flights_new = flights.groupby(['year', 'month'])['passengers'].agg(['sum', 'mean', 'max', 'min']).reset_index()
flights_new
```

```python
# grouped
grouped = flights.groupby(['year', 'month'])['passengers'].mean()
grouped
```

```python
# unstacked
grouped.unstack()
```

#### transform

```python
# 과일 데이터 로딩 
fruits = pd.read_csv('fruits.csv')
fruits
```

```python
# groupby() vs transform()
print(fruits.groupby('fruits')['price'].mean(), '\n')
print(fruits.groupby('fruits')['price'].transform('mean'))
```

```python
# 결측값을 각 과일의 평균가격으로 대체 
avg_price = fruits.groupby('fruits')['price'].transform('mean')
fruits['price'] = fruits['price'].fillna(avg_price)
fruits
```

#### 시계열 다루기 

```python
# 시계열 샘플 데이터 로딩 
df = pd.read_csv('datetime1.csv')
df.head(1).T
```

```python
# 날짜, 날짜/시간 데이터를 datetime 형태로 변경 
df['Date1'] = pd.to_datetime(df['Date1'])
df['Date2'] = pd.to_datetime(df['Date2'], format = '%Y:%m:%d')
df['Date3'] = pd.to_datetime(df['Date3'], format = '%y/%m/%d')
df['Date4'] = pd.to_datetime(df['Date4'], format = '%m/%d/%Y')
df['Date5'] = pd.to_datetime(df['Date5'])

df['DateTime1'] = pd.to_datetime(df['DateTime1'], format = '%y-%m-%d %H:%M:%S')
df['DateTime2'] = pd.to_datetime(df['DateTime2'], format = '%Y-%m-%d %H-%M-%S')
df['DateTime3'] = pd.to_datetime(df['DateTime3'], format = '%m/%d/%Y %I:%M:%S %p') #포맷 생략 가능 
df['DateTime4'] = pd.to_datetime(df['DateTime4'])

# 결과확인 
df.head(1).T
```

```python
df.info()
```

```python
# datetime -> year, month, hour ... 
df['year'] = df['DateTime1'].dt.year
df['month'] = df['DateTime1'].dt.month
df['day'] = df['DateTime1'].dt.day
df['hour'] = df['DateTime1'].dt.hour
df['minute'] = df['DateTime1'].dt.minute
df['second'] = df['DateTime1'].dt.second

df['day_of_week'] = df['DateTime1'].dt.day_of_week
df['day_name'] = df['DateTime1'].dt.day_name()

# 결과 확인 
df.head(1).T
```

```python
# datetime2 불러오기 
df2 = pd.read_csv('datetime2.csv')

# 연도-월 추출 
df2['yy_mm'] = df['DateTime4'].dt.to_period('M')
df2
```

```python
# 시간계산 
df2['+24hours'] = df['DateTime4'] + pd.Timedelta(hours=24)
df2['+100days'] = df['DateTime4'] + pd.Timedelta(days=100)
df2
```

```python
# 지점별 평균 배달소요시간 계산 예제 
## 딜리버리타임 데이터 로딩, 컬럼 포맷을 datetime으로 변경 
df = pd.read_csv('delivery_time.csv')

df['order_time'] = pd.to_datetime(df['order_time'])
df['delivery_time'] = pd.to_datetime(df['delivery_time'])

## 각각의 배달소요시간 계산 (분 단위)
df['duration'] = (df['delivery_time'] - df['order_time']).dt.total_seconds() / 60 

## groupby()로 지점별 평균 배달소요시간 계산 
df.groupby('store')['duration'].mean() 
```
